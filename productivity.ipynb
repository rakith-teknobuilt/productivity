# productivity based on work done percentage

method1

```python
dp=dp[['dailyProgressId', 'iwpId','iwpTotalQuantity', 'iwpUom', 'iwpToDoQuantity',
       'iwpDoneQuantity', 'iwpDonePercentage', 'iwpProgressDate', 'postDate']]

iwp=iwp[['id','cwpRefNo','iwpPlannedStartDate',
       'iwpPlannedComplete','iwpForecastStartDate',
       'iwpForecastComplete', 'iwpActualStartDate',
       'iwpActualComplete']]
```

```python
result = dp.merge(iwp, left_on='iwpId', right_on='id', how='left')
```

```python
result.columns
```

```python
result=result[['iwpId', 'iwpTotalQuantity', 'iwpUom',
       'iwpDonePercentage','iwpProgressDate','cwpRefNo', 'iwpPlannedStartDate',
       'iwpPlannedComplete', 'iwpForecastStartDate', 'iwpForecastComplete',
       'iwpActualStartDate', 'iwpActualComplete']]
```

```python
result=result.dropna()
result
```

```python
result['iwpDonePercentage'] = result['iwpDonePercentage'].replace('%', '', regex=True)
result
```

```python
result.replace(r'\\N', np.nan, regex=True, inplace=True)

# Drop rows with any NaN values
result = result.dropna()


result['iwpDonePercentage'] = pd.to_numeric(result['iwpDonePercentage'], errors='coerce')
# Replace inf and -inf with NaN
result['iwpDonePercentage'].replace([np.inf, -np.inf], np.nan, inplace=True)

# Optionally, handle NaNs by filling them or dropping them
result['iwpDonePercentage']= result['iwpDonePercentage'].fillna(0)
result['iwpDonePercentage'] = result['iwpDonePercentage'].astype(int)
```

```python
result['iwpTotalQuantity'] = pd.to_numeric(result['iwpTotalQuantity'])

result_df = result.groupby('iwpId').agg(
    total_finishing_percentage=('iwpDonePercentage', 'sum'),
#     iwpPlannedStartDate=('iwpPlannedStartDate', 'min'),
#     iwpPlannedComplete=('iwpPlannedComplete', 'max'),
    iwpPlannedStartDate=('iwpForecastStartDate', 'min'),
    iwpPlannedComplete=('iwpForecastComplete', 'max'),
    iwpPlannedActualDate=('iwpActualStartDate', 'min'),
    iwpActualComplete=('iwpActualComplete', 'max')
    #iwpTotalQuantity=('iwpTotalQuantity', lambda x: x.sum() / len(x))
).reset_index()
```

```python
result_df.loc[result_df['total_finishing_percentage'] >= 100, 'total_finishing_percentage'] = 100
result_df
```

```python
# Convert date columns to datetime, handling errors
result_df['iwpPlannedStartDate'] = pd.to_datetime(result_df['iwpPlannedStartDate'], errors='coerce')
result_df['iwpPlannedComplete'] = pd.to_datetime(result_df['iwpPlannedComplete'], errors='coerce')
result_df['iwpPlannedActualDate'] = pd.to_datetime(result_df['iwpPlannedActualDate'], errors='coerce')
result_df['iwpActualComplete'] = pd.to_datetime(result_df['iwpActualComplete'], errors='coerce')

# Calculate durations
result_df['planned_duration'] = result_df['iwpPlannedComplete'] - result_df['iwpPlannedStartDate']
result_df['actual_duration'] = result_df['iwpActualComplete'] - result_df['iwpPlannedActualDate']
result_df = result_df.dropna()
result_df
```

```python
completed_df = result_df[result_df['total_finishing_percentage'] == 100]
inprogress_df = result_df[result_df['total_finishing_percentage'] < 100]
```

```python
# Ensure the hours columns are numeric
man_hour_cwp['cwpPlannedHours'] = pd.to_numeric(man_hour_cwp['cwpPlannedHours'], errors='coerce')
man_hour_cwp['cwpForecastHours'] = pd.to_numeric(man_hour_cwp['cwpForecastHours'], errors='coerce')
man_hour_cwp['cwpActualHours'] = pd.to_numeric(man_hour_cwp['cwpActualHours'], errors='coerce')

# Drop rows with 0 or negative values in the specified columns
man_hour_cwp = man_hour_cwp[
    (man_hour_cwp['cwpPlannedHours'] > 0) &
    (man_hour_cwp['cwpForecastHours'] > 0) &
    (man_hour_cwp['cwpActualHours'] > 0)
]
```

```python
cwp=cwp[['cwpNo','discipline_id','cwpPlannedRelease','cwpPlannedComplete','cwpForecastRelease','cwpForecastComplete','cwpActualRelease','cwpActualComplete']]
# Convert the relevant columns to datetime, invalid parsing will be set as NaT
cwp['cwpPlannedRelease'] = pd.to_datetime(cwp['cwpPlannedRelease'], format='%Y-%m-%d', errors='coerce')
cwp['cwpPlannedComplete'] = pd.to_datetime(cwp['cwpPlannedComplete'], format='%Y-%m-%d', errors='coerce')
cwp['cwpForecastRelease'] = pd.to_datetime(cwp['cwpForecastRelease'], format='%Y-%m-%d', errors='coerce')
cwp['cwpForecastComplete'] = pd.to_datetime(cwp['cwpForecastComplete'], format='%Y-%m-%d', errors='coerce')
cwp['cwpActualRelease'] = pd.to_datetime(cwp['cwpActualRelease'], format='%Y-%m-%d', errors='coerce')
cwp['cwpActualComplete'] = pd.to_datetime(cwp['cwpActualComplete'], format='%Y-%m-%d', errors='coerce')

cwp = cwp.dropna(subset=['cwpPlannedRelease','cwpPlannedComplete','cwpForecastRelease','cwpForecastComplete','cwpActualRelease','cwpActualComplete'])
cwp
```

```python
cwp['planned_difference'] = (cwp['cwpPlannedComplete'] - cwp['cwpPlannedRelease']).dt.days
cwp['forecast_difference'] = (cwp['cwpForecastComplete'] - cwp['cwpForecastRelease']).dt.days
cwp['actual_difference'] = (cwp['cwpActualComplete'] - cwp['cwpActualRelease']).dt.days
```

```python
cwp
```

```python
cwp = cwp[
    (cwp['planned_difference'] > 0) &
    (cwp['forecast_difference'] > 0) &
    (cwp['actual_difference'] > 0)
]

cwp
```

```python
print("Total_planned_days_for_cwp", sum(list(cwp['planned_difference'])))
print("Total_forecast_days_for_cwp", sum(list(cwp['forecast_difference'])))
print("Total_actual_days_for_cwp", sum(list(cwp['actual_difference'])))
```

```python
grouped_df = cwp_pe.groupby('discipline_id').agg({
    'forecast_difference': 'sum',
    'actual_difference': 'sum'
}).reset_index()

grouped_df['forecast_hours']=grouped_df['forecast_difference']*10
grouped_df['actual_hours']=grouped_df['actual_difference']*10
grouped_df
```

```python
grouped_df['performace_efficiency']=round(grouped_df['forecast_hours']/grouped_df['actual_hours'] ,2)
grouped_df
```

```python
dis = pd.read_csv('pace_workmanagement_discipline11.csv')
dis=dis[['disciplineId','disciplineName']]
dis = dis.rename(columns={'disciplineId': 'discipline_id'})
```

```python
result = pd.merge(grouped_df, dis, how='left', on='discipline_id')
result
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Melt the DataFrame to long format for Seaborn
df_melted = result[['disciplineName', 'forecast_hours', 'actual_hours']].melt(id_vars='disciplineName', var_name='Type', value_name='Hours')

# Plotting
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df_melted, x='disciplineName', y='Hours', hue='Type')

# Rotate and align x-axis labels
plt.xticks(rotation=45, ha='right')

# Adding titles and labels
plt.title('Forecast vs Actual Hours by Discipline Name')
plt.xlabel('Discipline Name')
plt.ylabel('Hours')

# Adding the number labels on each bar with a smaller font size
for container in ax.containers:
    ax.bar_label(container, fontsize=7)

plt.tight_layout()  # Adjust layout to make room for the rotated labels

plt.show()
```
